1. Use the matplotlib Object-Oriented method to create graph of function $g(x) = \frac{x}{\sqrt{1+x^2}}$ inside $f(x) = \frac{e^x}{1+e^x}$ as shown below. 

<img src="Files/Problem1.png" width="700">

Requirements:
- (5 pts) g(x) inside f(x)
- (5 pts) plot the function for 100 points across the range [-20,20]
- (2 pts) size of the figure should be 10 x 10 inches.
- (2 pts) different color for f(x) and g(x)
- (2 pts) xlabel, ylabel and title for f(x) and g(x)
- (2 pts) set proper line width and line style.
- (2 pts) save the figure to a "png" figure.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20,20,100)
f_x = np.exp(x)/(1+np.exp(x))
g_x = x/np.sqrt(1+x**2)

fig= plt.figure(figsize = (10,10))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.15,0.5,0.3,0.3])

ax1.plot(x,f_x,'-r',lw = 1)
ax1.set_xlabel('x for f(x)')
ax1.set_ylabel('sigmod function')
ax1.set_title(r'$f(x) = \frac{e^x}{1+e^x}$')

ax2.plot(x,f_x,'--c', lw =4)
ax2.set_xlabel('x for g(x)')
ax2.set_ylabel('g(x)')
ax2.set_title(r'$g(x) = \frac{x}{\sqrt{1+x^2}}$')

plt.savefig('problem1.png')
```

3. (20 pts) Create a scatter plot comparing the carat of the diamonds to their price. Label the x-axis as "Carat", the y-axis as "Price", and the title as "Price vs. Carat of Diamonds". What can you infer from this plot regarding the relationship between carat and price (No more than 2 sentences)?

```python
plt.figure(figsize=(10,6))
# Plotting the scatter plot
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.9, color='skyblue')

# Adding labels and title
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Price vs. Carat of Diamonds')
plt.grid(axis = 'y')
plt.text(3,7000, 'Outlier', size=15)

# Display the plot
plt.tight_layout()
plt.show()
plt.savefig('problem3.png');
```

    <b><p style="font-size: XX-large">Problem 4: Pie plot</p></b> </div>

4. (20 pts) Display a pie chart that represents the proportion of diamonds for each cut quality. Make sure to include a legend and percentages on the chart.

Rubic:
- (5 pts) Correct pie plot function
- (3 pts) Proper figure size
- (3 pts) Proper plot title
- (3 pts) Legend and proper legend location.
- (3 pts) Percentage on the chart with one decimal place.
- (3 pts) Using distinct colors for each slice (or wedge) of the pie chart

```python
cut_counts = diamonds['cut'].value_counts()
plt.figure(figsize = (5,5))
plt.pie(cut_counts, labels = cut_counts.index, autopct ='%1.1f%%', colors = plt.cm.tab20c.colors)
plt.legend(cut_counts.index, title="Cut Quality", loc="best", bbox_to_anchor =(1,0.9))

# Display the plot
plt.tight_layout()
plt.title('Proportion of Diamonds by Cut Quality')
plt.show();
```

5. (20 pts) Plot separate histograms of the price column for each clarity grading. Use subplots to arrange these histograms (bins =20) in a grid where there's just one column and the number of rows is equal to the number of unique clarity grades. 

Rubic:
- (10 pts) Correct Subplot function and data
- (2 pts) Set figure size and avoid overlap (hint: `plt.tight_layout()`)
- (2 pts) Customize the 'color' for the histogram
- (2 pts) Set the title of the subplot to display which clarity grade the histogram represents. e.g. Price Distribution for Clarity: SI2
- (2 pts) Label the x-axis and y-axis
- (2 pts) Check for efficiency. (hint: using a for loop to iterate over each unique clarity grade and its associated subplot (axis))

```python
# Get unique clarity grades
clarities = diamonds['clarity'].unique()

# Create a grid of subplots
fig, axes = plt.subplots(nrows=len(clarities), ncols=1, figsize = (10, 20))

# Plot separate histograms for each clarity grade
for ax, clarity in zip(axes, clarities):
    ax.hist(diamonds[diamonds['clarity'] == clarity]['price'], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f'Price Distribution for Clarity: {clarity}')
    ax.set_xlabel('Price')
    ax.set_ylabel('Count')
    
plt.tight_layout()
plt.show()
```
