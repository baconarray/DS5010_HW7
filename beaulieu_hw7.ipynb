1. Use the matplotlib Object-Oriented method to create graph of function $g(x) = \frac{x}{\sqrt{1+x^2}}$ inside $f(x) = \frac{e^x}{1+e^x}$ as shown below. 

<img src="Files/Problem1.png" width="700">

Requirements:
- (5 pts) g(x) inside f(x)
- (5 pts) plot the function for 100 points across the range [-20,20]
- (2 pts) size of the figure should be 10 x 10 inches.
- (2 pts) different color for f(x) and g(x)
- (2 pts) xlabel, ylabel and title for f(x) and g(x)
- (2 pts) set proper line width and line style.
- (2 pts) save the figure to a "png" figure.

```python
import numpy as np
import matplotlib.pyplot as plt

#Problem 1

x = np.linspace(-20,20,100)
y1 = (np.e ** x)/(1 + (np.e ** x))
y2 = x / ((1 + (x**2))**0.5)

fig = plt.figure(figsize=(10,10))

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])#lbwh
ax2 = fig.add_axes([0.2,0.5,0.25,0.2])

ax1.plot(x,y1, 'r')
ax1.set_title('$f(x) = \\frac{e^x}{1+e^x}$')
ax1.set_xlabel('$x$ for $f(x)$')
ax1.set_ylabel('Sigmond Function')

ax2.plot(x,y2, '--c', lw=3)
ax2.set_title('$g(x) = \\frac{x}{\\sqrt{1+x^2}}$')
ax2.set_xlabel('$x$ for $g(x)$')
ax2.set_ylabel('$g(x)$')

plt.savefig('problem1.png')
```

5. (20 pts) Plot separate histograms of the price column for each clarity grading. Use subplots to arrange these histograms (bins =20) in a grid where there's just one column and the number of rows is equal to the number of unique clarity grades. 

Rubic:
- (10 pts) Correct Subplot function and data
- (2 pts) Set figure size and avoid overlap (hint: `plt.tight_layout()`)
- (2 pts) Customize the 'color' for the histogram
- (2 pts) Set the title of the subplot to display which clarity grade the histogram represents. e.g. Price Distribution for Clarity: SI2
- (2 pts) Label the x-axis and y-axis
- (2 pts) Check for efficiency. (hint: using a for loop to iterate over each unique clarity grade and its associated subplot (axis))

```python
#Problem 5

clarity = diamonds['clarity'].unique()

fig, axs = plt.subplots(nrows=len(clarity),ncols=1,figsize=(4,18),sharey=True)

for i in range(len(clarity)):
    diamond_bool = diamonds['clarity'] == clarity[i]
    prices = diamonds[diamond_bool == True]['price']
    axs[i].hist(prices,bins=20,color='xkcd:royal blue')
    axs[i].set_title("Price Distribution for Clarity: "+clarity[i])
    axs[i].set_xlabel("Price")
    axs[i].set_ylabel("#")

plt.tight_layout()

plt.savefig("problem5.png");
```
